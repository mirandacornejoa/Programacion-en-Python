```python
data = pd.read_csv('adult.data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?**

```python
data['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
data[data['sex'] == 'Female']['age'].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
gc = data['native-country'] == 'Germany'
gp = (gc.sum() / data.shape[0]) * 100
print(gp)
```

4-5. __What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year?__

```python
# mean and standard deviation of age for those who earn more than 50K per year (salary feature)
mean = data[data["salary"] == ">50K"]["age"].mean()
std = data[data["salary"] == ">50K"]["age"].std()
print("mean= ",mean)
print("std= ",std)
```

```python
# mean and standard deviation of age for those who earn less than 50k per year (salary feature)
mean2 = data[data["salary"] == "<=50K"]["age"].mean()
std2 = data[data["salary"] == "<=50K"]["age"].std()
print("mean= ",mean2)
print("std= ",std2)
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education â€“ Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
education = ["Bachelors", "Prof-school", "Assoc-acdm", "Assoc-voc", "Masters", "Doctorate"]
heducation = data[data["salary"] == ">50K"]["education"].isin(education).all()
print(heducation)
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
race_sex = data.groupby(["race", "sex"])["age"].describe()
print(race_sex)

amereskimo = data[(data["race"] == "Amer-Indian-Eskimo") & (data["sex"] == "Male")]["age"].max()
print("Max age of men of Amer-Indian-Eskimo race: " , amereskimo)
```

**8. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
maxh = data["hours-per-week"].max()
print("Max hours per week: ", maxh)

num_maxh = data[data["hours-per-week"] == maxh].shape[0]
print("Number of people working max hours: ", num_maxh)

perc_max_hours = (data[(data["hours-per-week"] == maxh) & (data["salary"] == ">50K")].shape[0] / num_maxh) * 100
print("Percentage of those who earn a lot (>50K) among them: ", perc_max_hours)
```

**9. Find the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python
from datetime import time
time_country = data.groupby(["native-country", "salary"])["hours-per-week"].mean()
print(time_country)

time_japan = data[data["native-country"] == "Japan"].groupby(["salary"])["hours-per-week"].mean()
print(time_japan)
```
